Import library pada google collab
**cv** untuk Computer Vision, **numpy** untuk library phyton untuk operasi array, **matplotlib** untuk visualisasi data (diagram, plot, dll)

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
# inisialisasi cascade untuk object detection
cascade = cv.CascadeClassifier('/content/drive/MyDrive/Images/Project/haarcascade.xml')

# read dataset sebagai variabel brain dan brain 1
brain = cv.imread('/content/drive/MyDrive/Images/Project/Brain_Tumor/Y169.jpg')
brain1 = cv.imread('/content/drive/MyDrive/Images/Project/Brain_Tumor/Y169.jpg')

# mendeteksi objek dengan cascade untuk variabel brain
roi = cascade.detectMultiScale(brain)
roi1 = cascade.detectMultiScale(brain)

# Membuat figur subplot koordinat x dan y dari sudut kiri atas, serta lebar (w) dan tinggi (h) dengan library rectangle dengan ukuran 10x10
for(x,y,w,h) in roi:
  cv.rectangle(brain,(x,y),(x+w,y+h),(0,255,0),3)
plt.figure(figsize = (10,10))

# Mencetak diagram dengan library matplotlib dengan ketentuan format warna BG
f, axarr = plt. subplots(1,2,figsize=(15,5))
plt.subplot(121),plt.imshow(cv.cvtColor(brain1,cv.COLOR_BGR2RGB))

axarr[0].imshow(cv.cvtColor(brain1, cv.COLOR_BGR2RGB))
plt.subplot(122),plt.imshow(cv.cvtColor(brain,cv.COLOR_BGR2RGB))
axarr[1].imshow(cv.cvtColor(brain, cv.COLOR_BGR2RGB))

# Menampilkan hasil plot
plt.show()
```
